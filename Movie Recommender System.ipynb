###### Formatting Columns

```python
import ast
# converting genre and keywords
"""returning all genre and keywords"""
def convert(obj):
    lst = []
    for i in ast.literal_eval(obj):
        lst.append(i["name"])
    return lst
```

```python
movies["genres"] = movies["genres"].apply(convert)
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
# converting cast
"""returning top 3 cast"""
def convert_top3(obj):
    lst = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter <3:
            lst.append(i["name"])
        counter +=1
    return lst
```

```python
movies["cast"] = movies["cast"].apply(convert_top3)
```

```python
# converting crew
"""fetching director"""
def fetch_director(obj):
    lst = []
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            lst.append(i["name"])
            break
    return lst
```

```python
movies["crew"] = movies.crew.apply(fetch_director)
```

```python
# converting overview as list (for easy concatenation) with other columns
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# removing whitespace to avoid confusion during searching e.g. in Sam Worthington and Sam Mendes
movies['genres'] = movies.genres.apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies.keywords.apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies.cast.apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies.crew.apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] =  movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
df = movies[['movie_id', 'title', 'tags']]
df.head(1)
```

```python
new_df = df.copy()
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

### Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# function to stem the tags
def stemmer(text):
    lst = []
    
    for i in text.split():
        lst.append(ps.stem(i))
        
    return " ".join(lst)
```

```python
new_df['tags'] = new_df['tags'].apply(stemmer)
new_df.head(5)
```

```python
# calculate cosine distance similarity between movie vectors (to find out how similar or different two movies are)
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0] # it gives similarity of first movie with every other movie
```

**NOTE:** Similarity of first movie with first movie, second movie with second movie and so on will always be 1. Therefore the diagonal of **similarity matrix** will always be 1.

### Recommender System

```python
def recommender_sys(movie):
    movie_loc = new_df.index[new_df['title']==movie][0]
    similar_movies = similarity[movie_loc]
    similar_movies = sorted(list(enumerate(similar_movies)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in similar_movies:
        print(new_df.iloc[i[0]].title)
        print(new_df.iloc[i[0]].movie_id)
```

**Note** <br>
similar_movies = sorted(list(enumerate(similar_movies)), reverse=True, key=lambda x:x[1])[1:6]

This line sorts movie similarity scores and selects the top 5 recommendations (excluding itself).  

1. `enumerate(similar_movies)`: Pairs indices with similarity scores 
2. `list(...)`: Converts the enumerated object into a list of tuples `(index, similarity_score)`.  
3. `sorted(..., reverse=True, key=lambda x: x[1])`: Sorts in descending order based on similarity scores.  
4. `[1:6]`: Skips the first (self) and selects the next 5 most similar movies.  

*Note*:
- **`key=lambda x: x[1]`**: This specifies a function to determine the sorting criteria.
  - **`lambda x: x[1]`**: A lambda function that takes each tuple `x` (which consists of `(index, similarity_score)`) and uses `x[1]` (the similarity score) for sorting.
  
This means the list is sorted based on the similarity scores in descending order.
